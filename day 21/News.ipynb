```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from urllib.parse import urljoin
import time

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("user-agent=Mozilla/5.0")

s=Service('chromedriver-win64/chromedriver.exe')
driver = webdriver.Chrome(service=s, options=chrome_options)

# Setup your driver here
# Assuming you have the correct driver initialized (like chromedriver)
all_news_data = []
page_counts = 0
max_pages = 5

current_page_url = 'https://news.ycombinator.com/'
baseurl = 'https://news.ycombinator.com/'

while current_page_url and page_counts < max_pages:
    page_counts += 1
    driver.get(current_page_url)

    # Wait for the page to load
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'athing')))

    print(f"Page {page_counts} loaded")

    # Extract all news items
    news = driver.find_elements(By.CLASS_NAME, 'athing')

    for n in news:
        text = n.find_element(By.CLASS_NAME, 'titleline').text
        # print(f"-- {text}")
        all_news_data.append(text)

    try:
        # Find the 'more' button (if it exists)
        more_button = driver.find_element(By.CLASS_NAME, 'morelink')
        next_page_url = more_button.get_attribute('href')
        print('Found next page button')

        # Update the current page URL
        current_page_url = urljoin(baseurl, next_page_url)
        time.sleep(1)  # Sleep to avoid hammering the server

    except Exception as e:
        print("No more pages or error:", e)
        current_page_url = None  # Stop the loop if no more pages

driver.quit()
    
for i in all_news_data:
    print(i)
    
print("\nBrowser closed!!")
```
