```python

import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL="https://jeevee.com/"
headers={
    "User-Agent":"My web Scrapper 1.0-for educational purposes"

}
response=requests.get(URL,headers=headers)
response
```

```python
soup=BeautifulSoup(response.text,'lxml')
```

```python
print(soup.prettify())
```

```python
items=soup.find_all('div',class_="productDesc")
print(items)
```

```python
daraz_items=[]
for item in items:
    print(item)
    title=item.find('a',class_="app-shadow lg:aspect-[212/384] product-size mr-2 app-rounded relative shrink-0 cursor-pointer overflow-hidden bg-white products_productCard__TgLt6 border")
    price=item.find('span',class_="price")
    print(title)
    print(price)
    item_title=title.text
    item_detail={
        'title':item_title
    }
    daraz_items.append(item_detail)
print(daraz_items)
```
